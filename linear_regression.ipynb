```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
weather_data = pd.read_csv("dataset/Weather.csv", low_memory=False)

X=weather_data['MaxTemp'].values.reshape(-1,1)
Y=weather_data['MinTemp'].values.reshape(-1,1)
min_max_plot=plt
min_max_plot.scatter(X, Y, marker="o")
min_max_plot.title("Plot between min and max tempreture.")
min_max_plot.xlabel("MaxTemp")
min_max_plot.ylabel("MinTemp")
min_max_plot.show()

X_train, x_test, Y_train, y_test=train_test_split(X, Y, test_size=0.3)
linear_regression=LinearRegression()
linear_regression.fit(X_train, Y_train)
y_predict = linear_regression.predict(x_test)

min_max_train_plot=plt
min_max_train_plot.scatter(x_test, y_test, marker="o", color="r")
min_max_train_plot.title("Train Model Plot between min and max tempreture.")
min_max_train_plot.plot(x_test, y_predict, color='b')
min_max_train_plot.xlabel("MaxTemp")
min_max_train_plot.ylabel("MinTemp")
min_max_train_plot.show()
print("Regression Intercept", linear_regression.intercept_)
print("Regression Coefficient", linear_regression.coef_)
```
